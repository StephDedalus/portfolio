```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from matplotlib.animation import FuncAnimation
```

```python
df = pd.read_csv('netflix price.csv', index_col=0)  
df.head()
```

```python
sns.set_style("whitegrid")
bar = plt.figure(figsize=(20,10))
plt.title("Total Library Size", fontsize = 40)
sns.barplot(x=df.index, y=df['Total Library Size'])
#sns.color_palette("mako", as_cmap=True)
plt.xticks(rotation=90, fontsize=15)
plt.yticks(np.arange(0,8000,500), fontsize = 15)
plt.xlabel('Country',fontsize = 20)
plt.ylabel('Library Size',fontsize = 20 )


plt.savefig('total.png')
```

```python
sns.set_style("whitegrid")
bar = plt.figure(figsize=(20,10))
plt.title("No. of TV Shows", fontsize = 40)
sns.barplot(x=df.index, y=df['No. of TV Shows'])
plt.xticks(rotation=90, fontsize=15)
plt.yticks(fontsize = 15)
plt.xlabel('Country',fontsize = 20)
plt.ylabel('TV Shows',fontsize = 20 )

plt.savefig('tvshows.png')
```

```python
sns.set_style("whitegrid")
bar = plt.figure(figsize=(20,10))
plt.title("Cost Per Month - Basic ($)", fontsize = 40)
sns.barplot(x=df.index, y=df['Cost Per Month - Basic ($)'], palette ="rocket")
#sns.color_palette("mako", as_cmap=True)
plt.xticks(rotation=90, fontsize=15)
plt.yticks(fontsize = 15)
plt.xlabel('Country',fontsize = 20)
plt.ylabel('Cost Per Month - Basic ($)',fontsize = 20 )

plt.show()
plt.savefig('cost.png')
```

```python
sns.set_style("whitegrid")
bar = plt.figure(figsize=(20,10))
plt.title("Cost Per Month - Standard ($)", fontsize = 40)
sns.barplot(x=df.index, y=df['Cost Per Month - Standard ($)'], color='xkcd:sky blue')
#sns.color_palette("mako", as_cmap=True)
plt.xticks(rotation=90, fontsize=15)
plt.yticks(fontsize = 15)
plt.xlabel('Country',fontsize = 20)
plt.ylabel('Cost Per Month - Standard ($)',fontsize = 20 )


plt.savefig('standard.png')
```

```python
sns.set_style("whitegrid")
bar = plt.figure(figsize=(20,10))
plt.title("Cost Per Month - Premium ($)", fontsize = 40)
sns.barplot(x=df.index, y=df['Cost Per Month - Premium ($)'])
#sns.color_palette("mako", as_cmap=True)
plt.xticks(rotation=90, fontsize=15)
plt.yticks(np.arange(0,30,1),fontsize = 15)
plt.xlabel('Country',fontsize = 20)
plt.ylabel('Cost Per Month - Premium ($)',fontsize = 20 )

plt.show()
plt.savefig('month.png', bbox_inches='tight')
```

```python
df = pd.read_csv('netflix price.csv', index_col=None)  
df.head()
```

```python
f, ax = plt.subplots(figsize=(4,14))

cost_basic_sort = df.sort_values("Cost Per Month - Premium ($)", ascending=False)


sns.set_color_codes("pastel")
sns.barplot(x="Cost Per Month - Premium ($)", y='Country', data=cost_basic_sort,
            label="Premium", color="r")
    

sns.barplot(x="Cost Per Month - Standard ($)", y="Country", data=cost_basic_sort,
            label="Standard", color="b")


sns.barplot(x="Cost Per Month - Basic ($)", y="Country", data=cost_basic_sort,
            label="Basic", color="g")


ax.legend(bbox_to_anchor=(1.05, 0.3), loc='upper left', borderaxespad=0.) # (x, y) = (1.05, 0.3)
ax.set(title="Netflix Subscription Cost Per Month in Different Countries",
       xlabel="Dollars ($)", xlim=(0,30))
sns.despine(left=True, bottom=True)
plt.savefig('total.png', bbox_inches='tight')
```

```python
f, ax = plt.subplots(figsize=(4,6))


cost_basic_sort = df.sort_values("Cost Per Month - Premium ($)", ascending=False).head(10)


sns.set_color_codes("pastel")
sns.barplot(x="Cost Per Month - Premium ($)", y="Country", data=cost_basic_sort,
            label="Premium", color="r")
    

sns.barplot(x="Cost Per Month - Standard ($)", y="Country", data=cost_basic_sort,
            label="Standard", color="b")


sns.barplot(x="Cost Per Month - Basic ($)", y="Country", data=cost_basic_sort,
            label="Basic", color="g")


ax.legend(bbox_to_anchor=(1.05, 0.3), loc='upper left', borderaxespad=0.) # (x, y) = (1.05, 0.3)
ax.set(title="Netflix Subscription Cost Per Month \n in Different Countries (Top 10)",
       xlabel="Dollars ($)",)
sns.despine(left=True, bottom=True)
```

```python
num_columns = df.select_dtypes(include=np.number).columns

matrix = np.triu(df[num_columns].corr())
f, ax = plt.subplots(figsize=(10,8))
ax = sns.heatmap(df[num_columns].corr(), annot=True, mask=matrix, vmax=1, vmin=0)
ax.set(title="Correlation Matrix")
```
